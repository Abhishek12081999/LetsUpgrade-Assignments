# Task 22
### Take a sample of size 50 from the column damageDealt for 100 times and calculate its mean. Plot it on a histogram and comment on its distribution.

```python
sample_damage = pd.DataFrame()
```

```python
damage = pd.DataFrame(pubg_data['damageDealt'])
for i in range(0,100):
    sample_pubg = pd.DataFrame(damage.sample(50))
    sample_damage = sample_damage.append(sample_pubg)
sample_damage
```

```python
sample_damage['damageDealt'].mean()
```

```python
%matplotlib inline
sns.set(rc={'figure.figsize':(11,8)}, font_scale=1.5, style='whitegrid')
sns.distplot( sample_damage['damageDealt'] )
plt.show()
```

***Looking at the data distribution, samples picked out has 0-1250(approx) Damage Dealt by the player. Maximum players have managed to not endure much damage***
